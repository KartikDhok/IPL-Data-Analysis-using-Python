```python
#mat['season']=pd.DatetimeIndex(mat['date']).year  #two extract the season 
run_match=ball.groupby(['BattingTeam'])['total_run'].sum().reset_index()

run_match.plot(kind='bar',x='BattingTeam',y='total_run',figsize=(12,12),color = "#4CAF50")
```

```python
batting_match=ball.groupby(['BattingTeam','batter'])['total_run'].sum().reset_index()
batting_top=batting_match.sort_values(by=['total_run'], ascending=False).head(10)
batting_top
```

```python
bowling_match=ball[ball['isWicketDelivery']!='runout' ]
bowling_match2=bowling_match.groupby(['bowler'])['isWicketDelivery'].sum().reset_index()
bow3=bowling_match2.rename(columns={"isWicketDelivery":"Total_Wickets"})
bow4=bow3.sort_values(by='Total_Wickets',ascending=False).head(10)
bow4.plot(kind='bar',x='bowler',y='Total_Wickets',figsize=(12,12),color = "#4CAF50")
print(bow4.iloc[:,:])

```
